**Question 2:** Are all CpGs associated with age? if not, which CpGs are more important for age prediction?

- The below code shows the shape of the selected columns of methylation values based on the given threshold requirement.

```python
from sklearn.feature_selection import VarianceThreshold

# Set the threshold for variance (can adjust this value)
threshold_value = 0.01

# Create the VarianceThreshold object
selector = VarianceThreshold(threshold=threshold_value)

# Fit the selector to the data (excluding the non-numeric columns if present)
numeric_cols = df_methyl_T.select_dtypes(include='number').columns
df_methyl_numeric = df_methyl_T[numeric_cols]

selector.fit(df_methyl_numeric)

# Get the columns that are selected (important) by the VarianceThreshold
selected_columns = df_methyl_numeric.columns[selector.get_support()]

# Create the DataFrame with only the selected columns
df_methyl_selected = df_methyl_numeric[selected_columns]
```

```python
df_methyl_selected = df_methyl_selected.T
df_methyl_selected
```
