###MethylDataWithDbgapId DataFrame

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python

file_path = "/content/metRmOlWithDbgapId.txt"
df_methyl = pd.read_csv(file_path, sep='\t', index_col=0)
# df_methyl = df_methyl.T
# print(df_methyl)
df_methyl.head()
```

```python
df_methyl
```

```python
# df_methyl.isnull().sum()
df_methyl.isnull().sum().sum()
```

```python
# Check for duplicate entries
duplicates = df_methyl.duplicated().sum()
print("Number of duplicate entries: ",duplicates)
```

```python
import matplotlib.pyplot as plt

# Define the individual's dbGaP ID (e.g., 34)
individual_id = '34'

# Extract the data for the specified individual
individual_data = df_methyl[individual_id]

# Get the CpG site names (row names)
cpg_sites = df_methyl.index

# Plot the CpG methylation values for the individual
plt.figure(figsize=(10, 6))
plt.plot(cpg_sites, individual_data, marker='o', linestyle='-')
plt.xlabel('CpG Sites')
plt.ylabel('Methylation Value')
plt.title(f'CpG Methylation Values for Individual {individual_id}')
plt.xticks(rotation=90)
plt.grid(True, axis='y')
plt.show()
```

```python
# Plot the CpG methylation values for the individual using a bar plot
plt.figure(figsize=(10, 6))
plt.bar(cpg_sites, individual_data)
plt.xlabel('CpG Sites')
plt.ylabel('Methylation Value')
plt.title(f'CpG Methylation Values for Individual {individual_id}')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```
